### **3. Create DAX Measures**

Go to **Modeling â†’ New Measure** and add:

```DAX
Total Employees = COUNTROWS('Table 1')

Attrition Count = CALCULATE(COUNTROWS('Table 1'), 'Table 1'[Attrition] = "Yes")

Active Employees = [Total Employees] - [Attrition Count]

Attrition Rate = DIVIDE([Attrition Count], [Total Employees], 0)

Average Age = AVERAGE('Table 1'[Age])

Average Monthly Income = AVERAGE('Table 1'[MonthlyIncome])
```
